## Histplot

## Countplot

```python
sns.countplot(data=df, x='Orders')
```

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
```

```python
A=sns.countplot(data=df, x='Zone', hue='Gender')
for bars in A.containers:
    A.bar_label(bars)
```

```python
plt.figure(figsize=(8,5))
plt.title('Age Group')
plt.ylim(0,3500)
plt.xlabel('Age Group')
sns.countplot(data=df, x='Age Group', hue='Gender');
```
